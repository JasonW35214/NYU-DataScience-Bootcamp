**1. Create a line plot of ZN and INDUS in the housing data.**

	a. For ZN, use a solid green line. For INDUS, use a blue dashed line.

	b. Change the figure size to a width of 12 and height of 8.

	c. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.

```python
import pandas as pd
import matplotlib.pyplot as plt
housing_csv = 'https://raw.githubusercontent.com/kartikjindgar/NYU-DataScience-Bootcamp-Fall23/main/Week4/data/boston_housing_data.csv'
housing = pd.read_csv(housing_csv)
#housing.head(6)

plt.style.use('bmh')
%matplotlib inline
housing[['ZN', 'INDUS']].plot(figsize=(12, 8), color={'ZN': 'green', 'INDUS': 'blue'}, style={'ZN': '-', 'INDUS': '--'})
```

**2. Create a bar chart using col1 and col2 of dummy data.**

	a. Give the plot a large title of your choosing.

	b. Move the legend to the lower-left corner.

	c. Do the same thing but with horizontal bars.

	d. Move the legend to the upper-right corner.

```python
import numpy as np
dummy_data = pd.DataFrame(np.random.randn(10, 4),
                  columns=['col1', 'col2', 'col3', 'col4'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])

dummy_data_ax = dummy_data[['col1', 'col2']].plot(kind='bar')
dummy_data_ax.set_title("col1 and col2 Vertical Bar Graph", fontsize=21, y=1)
dummy_data_ax.legend(loc=3)

dummy_data_ax2 = dummy_data[['col1', 'col2']].plot(kind='barh')
dummy_data_ax2.set_title("col1 and col2 Horizontal Bar Graph", fontsize=21, y=1)
dummy_data_ax2.legend(loc=1)
```
