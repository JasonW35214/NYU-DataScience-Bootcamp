1. Display Fibonacci Series upto 10 terms

```python
def fib(terms):
  """
  Generates and returns the first 'terms' terms of a Fibonacci Series in a list.
  """
  series = []
  for term in range(terms):
    if term == 0 or term == 1:
      series.append(term)
    else:
      series.append(series[term-2] + series[term-1])
  return series

print(fib(10))
```

5. Write a function that takes a word as an argument and returns the number of vowels in the word

```python
def num_vowels(word):
  """
  Takes a word as an argument and returns the number of vowels in the word
  """
  vowels = ['a', 'e', 'i', 'o', 'u']
  count = 0
  chars = list(word)
  for char in chars:
    if char in vowels:
      count += 1
    else:
      pass
  return count

print(num_vowels("hello"))
```

7. Iterate from 1 to 15, printing whether the number is odd or even

```python
for number in range(1,15+1,1):
  if number % 2 == 0:
    print(number, ": even")
  else:
    print(number, ": odd")
```
